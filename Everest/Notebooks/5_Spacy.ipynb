```python
import spacy
from spacy import displacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")

doc = nlp("the quick brown fox jumps over the lazy dog")
print([token.text for token in doc])
```

```python
import pandas as pd 
pd.DataFrame({
    "Token": [token.text for token in doc],
    "Lemma": [token.lemma_ for token in doc],
    "POS": [token.pos_ for token in doc],
    "Tag": [token.tag_ for token in doc],
    "Dep": [token.dep_ for token in doc],
    "Shape": [token.shape_ for token in doc],
})
```

```python
import pandas as pd
import spacy

expeddf = pd.read_csv("himalayan_tidy/expeddf_tidy.csv", low_memory=False)
text_accidents = expeddf.ACCIDENTS

nlp_acc = spacy.load("en_core_web_sm")

expeddf["ACCIDENTS_lemmas"] = text_accidents.apply(
    lambda txt: [
        token.lemma_.lower()
        for token in nlp_acc(str(txt))
        if not token.is_stop and not token.is_punct and not token.is_space
    ] if pd.notna(txt) else []
)

print(expeddf[["ACCIDENTS", "ACCIDENTS_lemmas"]].head(10))
```
