As the data was so unreadable in this format we decided to follow the Tidy_Tuesday recommendation and ammended 2_Data_Collection to change from integers into strings for both expeddf and peaksdf. 

The next stage of the EDA was to create expedition success flags, derive metrics like success rate per peak or route, death rate, altitude bands, and boolean indicators for oxygen use and hired personnel.
- Aggregate with `groupby` for the questions in the readme: distribution of PSTATUS by HIMAL_FACTOR, mean HEIGHTM by HIMAL_FACTOR, distributions of HEIGHTM for OPEN vs not, success rates by ROUTE1–4, oxygen vs non‑oxygen success, termination reasons (weather vs technical), and hired vs non‑hired death/success rates.
- Store reusable aggregates in separate DataFrames (e.g., `status_by_range`, `route_success`, `reason_counts`) and cache intermediate results to CSV/Parquet for version control


```python
expeddf["MDEATHS"].value_counts()
```

```python
expeddf[["MDEATHS", "HDEATHS"]]
```

```python
succcess_cols = ["SUCCESS1", "SUCCESS2", "SUCCESS3", "SUCCESS4"]

#Any success on any route
expeddf["success_flag"]= expeddf[succcess_cols].any(axis=1)

#Deaths flags 
death_cols = ["MDEATHS", "HDEATHS"]
expeddf["death_flag"] = expeddf[death_cols].gt(0).any(axis=1)

#Oxygen and hired personel 
expeddf["oxygen_used_flag"] = expeddf["O2USED"] > 0
expeddf["non_oxygen_used_flag"] = expeddf["O2NONE"] > 0
expeddf["no_hired_personnel_flag"] = expeddf["NOHIRED"] > 0    

```

Altitude bands 

```python
bins = [0, 6000, 7000, 8000, 9000]
labels = ['<6000m', '6000-6999m', '7000-7999m', '8000m+']
expeddf['altitude_band'] = pd.cut(peaksdf['HEIGHTM'], bins=bins, labels=labels, right=False)
expeddf['altitude_band'].value_counts().sort_index()
```

Grouped aggretates based on the business objectives in the readme. 

The PSTATUS is a number referencing whether the peak is unknown (0), unclimbed(1) or climbed (2)" and the status_by_range shows conceptually for each mountain range and for each peak PSTATUS.

```python
status_by_range = (
    peaksdf.groupby(["HIMAL", "PSTATUS"])
    .size()
    .rename("count")
    .reset_index()
)

#Mean HEIGHTM by HIMAL
mean_height_by_range = (
    peaksdf.groupby("HIMAL")["HEIGHTM"]
    .mean()
    .rename("mean_HEIGHTM")
    .reset_index()
)

#Distributions of HEIGHTM for OPEN vs not
heightm_by_open = (
    peaksdf.groupby("OPEN")["HEIGHTM"]
    .describe()
    .reset_index()
)   

#Success rates by main route Route1
route1_success = (
    expeddf.groupby("ROUTE1")["success_flag"]
    .mean()
    .rename("success_rate_route1") 
    .reset_index()
)

#Oxygen vs non‑oxygen success
oxygen_success = (
    expeddf.groupby("oxygen_used_flag")["success_flag"]
    .mean()
    .rename("success_rate_oxygen")
    .reset_index()
)

#Termination reasons (weather vs technical)
termination_reason_counts = (
    expeddf[expeddf["TERMREASON"].str.contains("weather|technical", case=False, na=False)]
    ["TERMREASON"]
    .value_counts()
    .rename("termreason_counts")
    .reset_index()
)
```

Save some reusable aggregates

```python
from Pathlib import Path

Path("himalayan_aggregates").mkdir(parents=True, exist_ok=True)

status_by_range.to_csv("himalayan_aggregates/status_by_range.csv", index=False)
mean_height_by_range.to_csv("himalayan_aggregates/mean_height_by_range.csv", index=False)
heightm_by_open.to_csv("himalayan_aggregates/heightm_by_open.csv", index=False)
route1_success.to_csv("himalayan_aggregates/route1_success.csv", index=False)
oxygen_success.to_csv("himalayan_aggregates/oxygen_success.csv", index=False)   
```
