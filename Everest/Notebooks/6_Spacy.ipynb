```python
import pandas as pd
import spacy

expeddf = pd.read_csv("himalayan_tidy/expeddf_tidy.csv", low_memory=False)
text_accidents = expeddf.ACCIDENTS

nlp_acc = spacy.load("en_core_web_sm")

expeddf["ACCIDENTS_lemmas"] = text_accidents.apply(
    lambda txt: [
        token.lemma_.lower()
        for token in nlp_acc(str(txt))
        if not token.is_stop and not token.is_punct and not token.is_space
    ] if pd.notna(txt) else []
)

print(expeddf[["ACCIDENTS", "ACCIDENTS_lemmas"]].head(10))
```

Applied same logic to Routememo is it has the longest text

```python
import pandas as pd
import spacy

expeddf = pd.read_csv("himalayan_tidy/expeddf_tidy.csv", low_memory=False)
text_routememo = expeddf.ROUTEMEMO

nlp_acc = spacy.load("en_core_web_sm")

expeddf.loc[1: ,"ROUTEMEMO_lemmas"] = text_routememo.iloc[:10].apply(
    lambda txt: [
        token.lemma_.lower()
        for token in nlp_acc(str(txt))
        if not token.is_stop and not token.is_punct and not token.is_space
    ] if pd.notna(txt) else []
)

print(expeddf[["ROUTEMEMO", "ROUTEMEMO_lemmas"]].head(10))
```

```python
# NER - Named Entity Recognition
import spacy
nlp = spacy.load("en_core_web_sm")

text_routememo = expeddf.ROUTEMEMO

for txt in text_routememo.iloc[:10].dropna():
    doc = nlp(str(txt))
    print("TEXT:", str(txt)[:80], "...")
    for ent in doc.ents:
        print("  ", ent.text, ent.label_)
    
```
