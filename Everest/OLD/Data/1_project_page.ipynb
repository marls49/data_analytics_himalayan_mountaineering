```python
import pandas as pd

df = pd.read_csv('exped_tidy_utf8.csv')
df_original = df.copy()

#convert SMTDAYS to numeric, erros coerce to NaN (removes non numbers)
df['SMTDAYS'] = pd.to_numeric(df['SMTDAYS'], errors ='coerce')

#remove rows where SMTDAYS is NaN or 0
df_clean = df[(~df['SMTDAYS'].isna()) & (df['SMTDAYS']!=0)]

print(df_clean['SMTDAYS'])

comparison = df_original.compare(df_clean)
num_changed = comparison.notna().sum().sum()
print(f"Number of changed cells: {num_changed}")
```

2. Second test was all of the colonms in data set

Results: Number of changed cells: 47897 

```python
import pandas as pd

#Read the CSV file 
df = pd.read_csv('exped_tidy_utf8.csv')
df_original = df.copy()

#All columns to numeric, non-numeric becomes NaN
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Remove rows where value is NaN or 0 
df_clean = df_numeric.dropna()
df_clean = df_clean[(df_clean != 0).all(axis=1)]

#compare with original 
df_clean_aligned = df_clean.reindex(df_original.index)
comparison = df_original.compare(df_clean_aligned)
num_changed = comparison.notna().sum().sum()

print(f"Number of changed cells: {num_changed} ")
```
